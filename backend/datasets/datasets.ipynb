```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
df = pd.read_csv('skoda_cleaned.csv')
df1 = pd.read_csv('audi_cleaned.csv')
df2 = pd.read_csv('volkswagen_cleaned.csv')
```

```python
df['model'] = df['model'].replace({'3-Feb': '2/3', '5-Apr': '4/5'})
df1['model'] = df1['model'].replace({'3-Feb': '2/3', '5-Apr': '4/5'})
df2['model'] = df2['model'].replace({'3-Feb': '2/3', '5-Apr': '4/5'})

d3 = pd.concat([df1, df2, df])
```

```python
d3.to_csv('all_cleaned.csv', index=False)
```
